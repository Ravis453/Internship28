```python
# Write a python program to scrape the details of top publications from Google Scholar from https://scholar.google.com/citations?view_op=top_venues&hl=en i) Rank ii) Publication iii) h5-index iv) h5-median
import requests
from bs4 import BeautifulSoup
import pandas as pd
page =requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
```

```python
soup=BeautifulSoup(page.content)
```

```python
publication=[]
for i in soup.find_all('td',class_="gsc_mvt_t"):
    publication.append(i.text)

publication
```

```python
index=[]
for i in soup.find_all('td',class_="gsc_mvt_n"):
    index.append(i.text)

index=index[::2]
    
index
```

```python
med=[]
for i in soup.find_all('span',class_="gs_ibl gsc_mp_anchor"):
    med.append(i.text)
med
```

```python
print(len(publication),len(med),len(index))
data=pd.DataFrame({'Top Publications':publication,"H-5 Median":med,"H-5 Index":index})

data
```
